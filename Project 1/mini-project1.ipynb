```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

credit_card['bill_over_limit'] = ((credit_card['BILL_AMT1']+credit_card['BILL_AMT2']+credit_card['BILL_AMT3']+credit_card['BILL_AMT4']+credit_card['BILL_AMT5']+credit_card['BILL_AMT6'])/6)/credit_card['LIMIT_BAL']
credit_card['age_sq'] = credit_card['AGE'] ** 2
credit_card['avg_pay_score'] = (credit_card['PAY_0'] + credit_card['PAY_2'] + credit_card['PAY_3']+ credit_card['PAY_4']+ credit_card['PAY_5']+credit_card['PAY_6'])/6
# credit_card['sum_pay_over_bill'] = (credit_card['PAY_0'] + credit_card['PAY_2'] + credit_card['PAY_3']+ credit_card['PAY_4']+ credit_card['PAY_5']+credit_card['PAY_6'])/(credit_card['BILL_AMT1']+credit_card['BILL_AMT2']+credit_card['BILL_AMT3']+credit_card['BILL_AMT4']+credit_card['BILL_AMT5']+credit_card['BILL_AMT6'])

X = credit_card[['bill_over_limit', 'avg_pay_score','MARRIAGE']]
y = credit_card['default.payment.next.month']
X1 = sm.add_constant(X)

x_train, x_test, y_train, y_test = train_test_split(X1, y, test_size = 0.10, random_state = 0)
model = sm.Logit(y_train, x_train)

results = model.fit()
# results_robust = results.get_robustcov_results(cov_type = 'HC1') # hetroskedacitiy corrector
print(results.summary())
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score, mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

prob_pred = results.predict(x_test)
y_pred = [0 if x<0.25 else 1 for x in prob_pred]

mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error: {rmse}')

r2 = r2_score(y_test, y_pred)
print(f'R-Squared: {r2}')

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy is: {accuracy}')

classif_report = classification_report(y_test, y_pred)
print(f'Classification report is: \n {classif_report}')

cm = confusion_matrix(y_test, y_pred)
print(f'Confusion matrix is: \n {cm}')

plt.imshow(cm, cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.xlabel("Predicted label")
plt.ylabel("True label")

tick_marks = np.arange(2)
plt.xticks(tick_marks, ['pos', 'neg'])
plt.yticks(tick_marks, ['pos', 'neg'])

for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], horizontalalignment="center")

plt.show()

fpr, tpr, t = roc_curve(y_test, prob_pred)
auc = roc_auc_score(y_test, prob_pred)
plt.plot(fpr, tpr, label ='AUC'+str(auc))
plt.legend()
plt.show()
```

This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables:
- X1: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit.
- X2: Gender (1 = male; 2 = female).
- X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others).
- X4: Marital status (1 = married; 2 = single; 3 = others).
- X5: Age (year).
- X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: 
    - X6 = the repayment status in September, 2005; 
    - X7 = the repayment status in August, 2005; . . .;
    - X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.
- X12-X17: Amount of bill statement (NT dollar). 
    - X12 = amount of bill statement in September, 2005; 
    - X13 = amount of bill statement in August, 2005; . . .; 
    - X17 = amount of bill statement in April, 2005.
- X18-X23: Amount of previous payment (NT dollar). 
    - X18 = amount paid in September, 2005; 
    - X19 = amount paid in August, 2005; . . .;
    - X23 = amount paid in April, 2005.

```python
credit_card['avg_pay'] = credit_card[["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]].mean(axis=1)
credit_card['avg_bill_amt'] = credit_card[["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]].mean(axis=1)
credit_card['avg_pay_amt'] = credit_card[["PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]].mean(axis=1)
credit_card
```

```python
# 1=yes, 0=no
credit_card['default.payment.next.month'].unique()
```

```python
# Count the number of defaults
default_count = credit_card['default.payment.next.month'].sum()

# Count the total number of credit card holders
total_count = len(credit_card)

# Calculate the default rate as a percentage
default_rate = default_count / total_count * 100

print('Default rate: {:.2f}%'.format(default_rate))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

df = credit_card

# Split the data into train and test sets
X = df[['LIMIT_BAL','SEX', 'EDUCATION', 'MARRIAGE', 'AGE','avg_pay', 'avg_bill_amt', 'avg_pay_amt']]
y = df["default.payment.next.month"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict the default rate
y_pred = model.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

import matplotlib.pyplot as plt
# not sure of the labels == we might wanna change it hehe
plt.imshow(cm, cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.xlabel("Predicted label")
plt.ylabel("True label")

tick_marks = np.arange(2)
plt.xticks(tick_marks, ['pos', 'neg'])
plt.yticks(tick_marks, ['pos', 'neg'])

for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], horizontalalignment="center")

plt.show()
```
