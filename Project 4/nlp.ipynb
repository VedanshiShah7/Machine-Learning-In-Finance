```python
# getting the S&P 500 stock tickers
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
table = data[0]
table.head()
tickers = table['Symbol'].tolist()
tickers[:5]
```

```python
# stock price for last 7 days for each ticker
import yfinance as yf
import datetime

start = datetime.datetime.today() - datetime.timedelta(7)
end = datetime.datetime.today()

data = yf.download(tickers, start=start, end=end, interval='1d')
data.head()
```

```python
# get stock price
data['Close'].head()
```

```python
# get the stock returns for each day including last day
data['Close'].pct_change().head()
```
